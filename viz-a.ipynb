```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

## loading data from the CSV file for the World cities average internet prices. 2010 - 2020 

```python
internet_data = pd.read_csv("cities_internet_prices_historical.24-10-2021.csv")
internet_data
```

## View the top 20 `Cities` and there `internet price in 2020` by there `Region`

```python
sns.set_theme()
data1= internet_data.head(20)
sns.displot(x="Internet Price, 2015" , y= "Internet Price, 2020",data=data1,hue="Region")
```

## Find the 10 highest regions internet price in 2019 

```python
region_data = internet_data.set_index("Internet Price, 2019").sort_index(ascending=False).head(10)
region_data
sns.relplot(x="City" , y= "Internet Price, 2019",data=region_data,  hue="Region")
```

## Get the data for the internet price for all years in `United States of America`

```python
usa_data = internet_data.set_index("City")
new_data = usa_data[usa_data["Country"] == "United States of America"]
garph = sns.barplot(data=new_data.head(10))
```
